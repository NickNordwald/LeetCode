## Contains Duplicate - Difficulty Easy

Given an integer array `nums`, return `True` if any value appears more than once in the array, otherwise return `False`.

```python
from typing import List 

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        
        seen = set()
        for val in nums:
            if val in seen:
                return True
            seen.add(val)
        return False
```

```python
example = [1, 2, 3, 4, 1]
test = Solution()

print(test.hasDuplicate(example))
```

#### Comments
- complexity O(n)
- utilizes that hash tables take O(1) time to check if an key exists already
- could have been equivalently done using python dictionaries as they are also implemented using hast tables (i.e. the keys of the dict are hashed)
- could technically be done faster using the `length()` function, however since difficulty easy I felt this was the more intended solution

## Group Anagrams - Difficulty Medium

Given an array of strings `strs`, group the anagrams together. You can return the answer in any order.

```python
from collections import defaultdict
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        
        anagrams = defaultdict(list)
        for s in strs:

            char_count = [0] * 26
            for char in s:
                char_count[ord(char) - ord('a')] += 1
            
            anagrams[tuple(char_count)].append(s)
        
        return list(anagrams.values())
```

```python
example = ["eat","tea","tan","ate","nat","bat"]
test = Solution()

print(test.groupAnagrams(example))
```

## Comments
- complexity O(nk) where k is the word length and n is the number of words
- assumes that we get words in lowercase characters without special characters
- alternative approach might be to sort the words before using them as hashes however the complexity is a bit worse
- have to convert `char_count` to a `tuple` since `list` is not hashable
